```python
#personal temptationnn
df = sns.load_dataset('titanic')

#findind the number of men who survived
male_survived = df[(df['sex'] == 'male') & (df['survived'] == 1)].shape[0]
num_men = df[df['sex'] == 'male'].shape[0]   #since shape returns us a tuple(row,column) so shape[0] == num of rows i.e the size what we need

#finding the number od women who survived
women_survived = df[(df['sex'] == 'female') & (df['survived'] == 1)].shape[0]
num_women = df[df['sex'] == 'female'].shape[0]

percentage_men_survived = (male_survived/num_men) * 100
percentage_women_survived = (women_survived/num_women) * 100
print('percentage of men survived : ',percentage_men_survived)
print('percentage of women survived : ',percentage_women_survived)
```

##Correlation
This line calculates the correlation matrix of the DataFrame using the .corr() method from pandas.

* Correlation measures how much two variables change together.

* It ranges from -1 to +1:

* +1  strong positive correlation (both increase together)

* -1  strong negative correlation (one increases, other decreases)

* 0  no correlation

So correlation becomes a DataFrame where each cell [i][j] is the correlation between column i and column j.



```python
#constructing the heatrmap
plt.figure(figsize = (10,10))
sns.heatmap(correlation,cbar = True,square = True,fmt = '.1f',annot = True,annot_kws={'size' : 8},cmap = 'Blues')
```

###sns.heatmap(...)

Parameters and whta each pf them does	   

* correlation	The 2D data to visualize (your correlation matrix).
* cbar=True	Shows the color bar (legend) on the side of the heatmap.
* square=True	Makes each cell a square (instead of rectangles).
* fmt='.1f'	Formatting for the numbers in each cell  here, 1 decimal place (e.g. 0.7, -0.4).
* annot=True	Shows the actual numbers (correlation values) in each cell.
* annot_kws={'size': 8}	Sets the size of the text used for annotations inside the cells.
* cmap='Blues'	Chooses the color palette. 'Blues' gives you light to dark blue * shades depending on value. Other examples: 'coolwarm', 'viridis', 'magma'.
